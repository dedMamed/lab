```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/Video_Games.csv')
data.head()
```

```python
data.info()
```

```python
data = data.fillna(-1000)
data.head()
```

```python
data.info()
```

```python
y = data['JP_Sales']
x = data.drop(columns = ['JP_Sales', 'Developer', 'Name'])
x.head()
```

```python
y.head()
```

```python
x_ = pd.get_dummies(x)
x_.head()
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(x_, y, test_size = 0.2)
x_.shape, y.shape, x_train.shape, y_train.shape, x_test.shape, y_test.shape,
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x_train, y_train)
```

$$Mean \space Absolute \space Error = MAE = \frac{1}{n}\sum_{i=1}^n{\left|Y_i - \hat{Y}_i\right|}$$

```python
from sklearn.metrics import mean_absolute_error
y_pred = regr.predict(x_train)
print("MAE (train) = ", mean_absolute_error(y_train, y_pred))
```

```python
y_pred = regr.predict(x_test)
print("MAE (test) = ", mean_absolute_error(y_test, y_pred))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
x.shape, y.shape, x_train.shape, y_train.shape, x_test.shape, y_test.shape,
```

```python
from catboost import Pool, CatBoostRegressor
train_pool = Pool(x_train, y_train, cat_features = ['Platform', 'Genre', 'Publisher', 'Rating'])
test_pool = Pool(x_test, y_test, cat_features = ['Platform', 'Genre', 'Publisher', 'Rating'])
regr = CatBoostRegressor()
regr.fit(train_pool)
```

```python
y_pred = regr.predict(x_train)
print("MAE (train) = ", mean_absolute_error(y_train, y_pred))

y_pred = regr.predict(x_test)
print("MAE (test) = ", mean_absolute_error(y_test, y_pred))
```

```python
regr = CatBoostRegressor(max_depth = 5, iterations = 2000, learning_rate = 0.2)
regr.fit(train_pool)
```
